```python
import requests
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
key = "7558457878736b7933374f6544556b"
url = f"http://openapi.seoul.go.kr:8088/{key}/json/RealtimeCityAir/1/999/"

# MSRDT(날짜), MSRSTE_NM(지역 ), PM10(미세먼지), IDEX_NM(상태)


response = requests.get(url)
responseDic = response.json()
```

```python
data = responseDic["RealtimeCityAir"]["row"]
data
```

```python
# 전체 데이터
df = pd.DataFrame(data)
df.to_csv("weather_df.csv", encoding='utf-8-sig')
```
